```python
import numpy as np
import pandas as pd
```

**what is pandas?**
pandas is a fast powerful,flexible and easy to use open source data analysis and manipulation tool,build on top of the python programming language.

**pandas series** is like  a column in a table. it is a 1-d array holding data of any type.

```python
#string
country = ["pakistan","india","USA","japan"]

pd.Series(country)
```

```python
#integer
match_run = [32,423,23,12]
pd.Series(match_run)
```

```python
#coustom index
marks = [34,53,23,54,77]
subject = ['English','Computer','Math','urdu','physics']
pd.Series(marks,index=subject)
```

```python
#setting a name
pd.Series(marks,index=subject,name = 'Ahmad marks')
```

Series from dict

```python
marks = {
    'english' : 23,
    'math' : 59,
    'Urdu' :49,
    'physics' :89,
    'Computer' : 99
}
marks_series = pd.Series(marks,name="marks")
marks_series
```
